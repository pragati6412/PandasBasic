```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("googleplaystore.csv")
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
imputer.fit(df.iloc[:,:].values)
df.iloc[:,:]=imputer.transform(df.iloc[:,:].values)
```

```python
df=df.dropna()
```

```python
df
```

## Free apps in Art and design Category

```python
df.values
```

```python
count=0
for i in df.values:
    if (i[1]=="ART_AND_DESIGN" and i[6]=="Free"):
        count+=1
print("Free Apps in ART and Design category is :",count)
```

##                                             OR

```python
df1=df[df['Category']=="ART_AND_DESIGN"]
print(len(df1[df1['Type']=="Free"]))
```

## Apps in Art and Design with rating greater than or equal to 4.5

```python
count=0
for i in df.values:
    if (i[1]=="ART_AND_DESIGN" and i[2]>=4.5):
        count+=1
print("Apps in Art and Design category with rating greater than 4.5 is :",count)
```

## OR

```python
df1=df[df['Category']=="ART_AND_DESIGN"]
print(len(df1[df1['Rating']>=4.5]))
```

## Free Apps in family category with rating greater than 4.5

```python
count=0
for i in df.values:
    if (i[6]=="Free" and i[1]=="FAMILY" and i[2]>4.5):
        count+=1
print("Free Apps in Family category with rating greater than 4.5 is :",count)
```

## OR

```python
df1=df[df['Category']=="FAMILY"]
df2=df1[df1['Type']=="Free"]
df3=df2[df2['Rating']>4.5]
print(len(df3))
```
