APPS FOR EACH CONTENT RATING

```python
n1=len(data['Content Rating'].unique())
for i in range(n1):
    count1=0
    for j in data['Content Rating']:
        if j==(data['Content Rating'].unique())[i]:
            count1+=1
    print(f"No of Apps in {(data['Content Rating'].unique())[i]} category is {count1}")
```

No. of apps in each category

```python
categories={}
for name in data['Category'].unique():
    count=0
    for i in data['Category']:
        if i==name:
            count+=1
    categories[name]=count
```

```python
for key in categories:
    print(key,categories[key])
```

No. of apps in each Type

```python
types={}
for name in data['Type'].unique():
    count=0
    for i in data['Type']:
        if i==name:
            count+=1
    types[name]=count
```

```python
for i in types:
    print(i,types[i])
```
